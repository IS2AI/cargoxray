```python
from numpy import nan
import pandas as pd
from pathlib import Path
```

```python
annotations = pd.read_json('data/annotations.json',
                           orient='records')
annotations = annotations.set_index('id')
annotations
```

```python
images = pd.read_json('data/images.json',orient='records')

try:
    images = images.rename(columns={'id': 'image_id'})
except:
    pass
images = images.set_index('image_id')
images
```

```python
annotations = annotations.replace((
    'brokkoli',
    'clouthes',
    'equpment',
    'Equipment',
    'grapes',
    'motobike',
    'motorcycle',
    'nectarine',
    'pears',
    'scooters',
    'textiles',
    'tomates',
    'Household goods',
    'Lamps',
    'Car wheels',
    'Clothes',
    'Shoes',
    'Spare parts',
    'appliances',
    'car wheels ',
    'carweels',
    'carwheels',
    'cars',
    'equipment '
), (
    'broccoli',
    'clothes',
    'equipment',
    'equipment',
    'grape',
    'bike',
    'bike',
    'nectarin',
    'pear',
    'scooter',
    'textile',
    'tomato',
    'household goods',
    'lamps',
    'car wheels',
    'clothes',
    'shoes',
    'spare parts',
    'appliance',
    'car wheels',
    'car wheels',
    'car wheels',
    'car',
    'equipment'
))

annotations['label'].drop_duplicates()
```

```python
annotations = annotations.join(images, how='inner').dropna()
annotations
```

```python
annotations['image_id'].to_frame() \
    .drop_duplicates()
```

```python
images = annotations['image_id'].to_frame() \
    .drop_duplicates() \
    .merge(right=images,
            how='inner',
            on='image_id') \
    .set_index('image_id') \
    .sort_index()
images
```

```python
labels = annotations \
    .groupby('label') \
    .count()['image_id'] \
    .to_frame() \
    .sort_values('image_id', ascending=False) \
    .rename(columns={'image_id': 'count'}) \
    .reset_index()
labels.index.rename('label_id', inplace=True)

display(images, annotations, labels)
```
